```python
import pandas as pd

from bs4 import BeautifulSoup as soup

import requests as req

url = "https://www.youtube.com/feed/trending"

page = req.get(url)


con = page.content

man = soup(con,'html.parser')

for x in man.findAll("div",{"id" : "content"}):
    g = x.findAll("a",{"class":"yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink spf-link"})
    for i in g:
        print(i.text)
```

```python
url = "https://www.youtube.com/"

q = req.get(url)

s = q.content

d = soup(s,'html.parser')

for x in d.findAll("div",{"id":"content"}):
    g = x.findAll("div",{"class":"yt-lockup-content"})
    for p in g:
        o = p.findAll("a")
        for j in o:
            print(j.text)
```

```python
import pandas as pd

from bs4 import BeautifulSoup as soup

import requests as req

url = "https://www.webopedia.com/quick_ref/textmessageabbreviations.asp"

page = req.get(url)

x = soup(page.content,"html.parser")

a = x.findAll("div",{"id":"article_content"})

b = a[0].findAll("td")

for x in b:
    print(x.text)
```
